```python
def get_largest_for_column(df,column):
    columnStr = str(column)
    topTwo = df.nlargest(2,[columnStr])[columnStr]
    largest = topTwo.min()
    idx = topTwo.idxmin()
    return largest,idx


def get_largest_all_columns(df,df_highcos):    
    df_out = pd.DataFrame()
    for col in df.columns:
        largest,idx = get_largest_for_column(df,col)
        assigned_row = col
        match_column = "largestMatchIndex"
        measurement_column = "cosineSimilarity"
        value_column = "essay"
        df_out.at[assigned_row,match_column] = str(int(idx))
        df_out.at[assigned_row,measurement_column] = largest
        df_out.at[assigned_row,value_column] = df_highcos['essay'][idx]
        
    return df_out
```

```python
# largest, idx = get_largest_for_column(df,0)
df_out = get_largest_all_columns(df,df_highcos)
df_out
```
