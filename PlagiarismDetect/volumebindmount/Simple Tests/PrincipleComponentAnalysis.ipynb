```python
import numpy as np
from sklearn.decomposition import PCA
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
pca = PCA(n_components=2)
pca.fit(X)
PCA(n_components=2)
pca_exvar_rat = pca.explained_variance_ratio_
pca_sing_vals = pca.singular_values_

print("pca explained variance ratio is: ",pca_exvar_rat)
print("pca singular values: ",pca_sing_vals)
```

```python
print("the type of pca is: ",type(pca))
```

```python
pca_transform_array = pca.transform(X)
print(pca_transform_array)
print("the type of pca_transform_array is: ",type(pca_transform_array))
print("the shape is: ",pca_transform_array.shape)
print("the dtype is: ",pca_transform_array.dtype)
```

```python
import matplotlib.pyplot as plt

x = pca_transform_array[:,0]
y = pca_transform_array[:,1]

# plot
fig, (ax1,ax2) = plt.subplots(1,2)

ax1.scatter(X[:,0],X[:,1])
ax2.scatter(x,y)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X3d = np.array([[-1, -1, -1], [-2, -1, -2], [-3, -2, -3], [1, 1, 1], [2, 1, 2], [3, 2, 3]])
pca_prime = PCA(n_components=2)
pca_prime.fit(X3d)
PCA(n_components=2)
pca_exvar_rat_prime = pca_prime.explained_variance_ratio_
pca_sing_vals_prime = pca_prime.singular_values_
pca_transform_array_prime = pca_prime.transform(X3d)

x_prime = pca_transform_array_prime[:,0]
y_prime = pca_transform_array_prime[:,1]

# plots
fig_w = plt.figure()
ax_w = fig_w.add_subplot(projection='3d')
ax_w.scatter(X3d[:,0],X3d[:,1],X3d[:,2])

fig, ax = plt.subplots(1)
ax.scatter(x_prime,y_prime)
```
