```python
import pandas as pd
```

```python
df = pd.read_csv('https://people.sc.fsu.edu/~jburkardt/data/csv/zillow.csv')
# perform lambda function to replace " on every instance found
rm_quote = lambda x: x.replace('"', '')
# set columns equal to rm_quote lambda function
df = df.rename(columns=rm_quote)
df.head(3)
```

```python
df.head(10).style.set_properties(**{'background-color': 'black',                                                   
                                    'color': 'lawngreen',                       
                                    'border-color': 'white'})
```

```python
df.head(10).style.format({"Baths": "{:.2f}"},precision=1)\
                    .highlight_max(color='lightgreen')
```
