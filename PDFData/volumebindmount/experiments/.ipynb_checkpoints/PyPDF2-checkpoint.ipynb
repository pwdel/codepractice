```python
import PyPDF2

def loadPDFObject(fullFilePath):
    # creating a pdf file object 
    pdfFileObj = open(fullFilePath, 'rb')
    # creating a pdf reader object 
    pdfReaderObj = PyPDF2.PdfFileReader(pdfFileObj) 
    
    return pdfReaderObj

def getNumPages(pdfReaderObj):
    pdfReaderPages = pdfReaderObj.numPages
    
    return pdfReaderPages

def getAllText(pdfReaderObj):
    
    textString = ""
    
    for page in range(0, pdfReaderObj.numPages):
        pageObj = pdfReaderObj.getPage(page)
        currentPageText = pageObj.extractText()
        textString = textString + currentPageText + "\n"
    
    return textString

def replacemany(our_str, to_be_replaced:tuple, replace_with:str):
    for nextchar in to_be_replaced:
        our_str = our_str.replace(nextchar, replace_with)
    return our_str
```

```python
folderString = '/home/jovyan/work/experiments/'
fileString = '03uno.pdf'
fullFilePath = folderString + fileString

pdfReaderObjInst = loadPDFObject(fullFilePath)
pdfReaderPagesInt = getNumPages(pdfReaderObjInst)

textStringInst = getAllText(pdfReaderObjInst)

tbr = ('','®','‐','.','+','-',':',')','(',',','/')
rw = ''

print(replacemany(textStringInst,tbr,rw))
```
